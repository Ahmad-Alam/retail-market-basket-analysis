# Product Analysis Notebook

This notebook analyzes product performance, seasonal trends, and inventory insights
from the online retail dataset

```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import calendar
import warnings

# Plotting settings
warnings.filterwarnings('ignore')

# Load and prepare data
df = pd.read_csv('../data/online_retail_cleaned.csv')
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

## Top/Bottom Performing Products Analysis

```python
## Calculate product performance metrics
product_metrics = df[df['IsOrder']].groupby('StockCode').agg({
    'Description': 'first',  # Get product name
    'Quantity': 'sum',       # Total quantity sold
    'TotalAmount': 'sum',    # Total revenue
    'InvoiceNo': 'nunique'   # Number of transactions
}).round(2)

product_metrics['AvgOrderValue'] = (product_metrics['TotalAmount'] / 
                                  product_metrics['InvoiceNo']).round(2)

# Sort by revenue to get top/bottom products
top_revenue_products = product_metrics.nlargest(10, 'TotalAmount')
bottom_revenue_products = product_metrics.nsmallest(10, 'TotalAmount')

# Visualize top 10 products by revenue
plt.figure(figsize=(12, 6))
sns.barplot(data=top_revenue_products.reset_index(), 
            x='TotalAmount', 
            y='Description',
            palette='viridis')
plt.title('Top 10 Products by Revenue')
plt.xlabel('Total Revenue')
plt.tight_layout()
plt.show()
```

## Seasonal Analysis

```python
# Create monthly sales aggregation
monthly_sales = df[df['IsOrder']].groupby(['Year', 'Month']).agg({
    'TotalAmount': 'sum',
    'Quantity': 'sum'
}).reset_index()

monthly_sales['Date'] = pd.to_datetime(monthly_sales[['Year', 'Month']].assign(DAY=1))

# Plot monthly sales trends
plt.figure(figsize=(15, 6))
plt.plot(monthly_sales['Date'], monthly_sales['TotalAmount'], marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Date')
plt.ylabel('Total Revenue')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Product Category Analysis

```python
# Extract categories from descriptions (using first word as category)
df['Category'] = df['Description'].str.split().str[0]

# Calculate category metrics
category_metrics = df[df['IsOrder']].groupby('Category').agg({
    'TotalAmount': 'sum',
    'Quantity': 'sum',
    'InvoiceNo': 'nunique',
    'CustomerID': 'nunique'
}).round(2)

category_metrics['AvgOrderValue'] = (category_metrics['TotalAmount'] / 
                                   category_metrics['InvoiceNo']).round(2)

# Visualize category performance
plt.figure(figsize=(12, 6))
sns.barplot(data=category_metrics.nlargest(10, 'TotalAmount').reset_index(),
            x='TotalAmount',
            y='Category',
            palette='viridis')
plt.title('Top 10 Categories by Revenue')
plt.xlabel('Total Revenue')
plt.tight_layout()
plt.show()
```

## Inventory Analysis

```python
# Calculate product turnover metrics
inventory_metrics = df.groupby('StockCode').agg({
    'Description': 'first',
    'Quantity': ['sum', 'mean', 'std'],
    'InvoiceDate': ['min', 'max']
}).round(2)

inventory_metrics.columns = ['Description', 'TotalQuantity', 'AvgQuantity', 
                           'StdQuantity', 'FirstSale', 'LastSale']

# Calculate days between first and last sale
inventory_metrics['DaysActive'] = (inventory_metrics['LastSale'] - 
                                 inventory_metrics['FirstSale']).dt.days

# Calculate daily turnover rate
inventory_metrics['DailyTurnover'] = (inventory_metrics['TotalQuantity'] / 
                                    inventory_metrics['DaysActive']).round(2)

# Identify slow-moving items (bottom 10% by daily turnover)
slow_moving = inventory_metrics.nsmallest(
    int(len(inventory_metrics) * 0.1), 
    'DailyTurnover'
)
```

## Export Results and Print Summary

```python
# Export results
product_metrics.to_csv('../data/product_metrics.csv')
category_metrics.to_csv('../data/category_metrics.csv')
inventory_metrics.to_csv('../data/inventory_metrics.csv')

# Print summary statistics
print("Product Analysis Summary:")
print(f"Total number of products: {len(product_metrics)}")
print(f"Total revenue: ${product_metrics['TotalAmount'].sum():,.2f}")
print(f"Average revenue per product: ${product_metrics['TotalAmount'].mean():,.2f}")
print(f"Number of categories: {len(category_metrics)}")
print(f"Number of slow-moving items: {len(slow_moving)}")
```
