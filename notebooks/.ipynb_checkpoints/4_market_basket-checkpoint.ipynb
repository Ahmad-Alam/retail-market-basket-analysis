# Market Basket Analysis
Analyzing customer purchase patterns and product associations using the Apriori algorithm to discover frequent itemsets and generate association rules.

## Setup
Import necessary libraries for data manipulation, market basket analysis, and visualization.

```python
# Import required libraries
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')
```

## Data Loading and Initial Exploration
Load the cleaned retail dataset and display basic information about its structure and content.

```python
# Load and prepare the data
df = pd.read_csv('../data/online_retail_cleaned.csv')

# Basic data overview
print("Dataset Shape:", df.shape)
print("\nSample of first few rows:")
display(df.head())
print("\nData Info:")
display(df.info())
```

## Transaction Data Preparation
Transform the data into transaction format and create one-hot encoded representation for market basket analysis.

```python
# Create transaction data
# Group items by invoice to create transaction lists
transactions = df.groupby(['InvoiceNo'])['Description'].agg(list).reset_index()
print("\nSample transactions:")
display(transactions.head())

# Convert transactions to one-hot encoded format
te = TransactionEncoder()
te_ary = te.fit_transform(transactions['Description'])
df_encoded = pd.DataFrame(te_ary, columns=te.columns_)
```

## Frequent Itemset Generation
Apply Apriori algorithm to identify frequently occurring item combinations with minimum support of 1%.

```python
# Memory-efficient frequent itemset generation
# Increase min_support and add max_len parameter to manage memory usage
frequent_itemsets = apriori(
    df_encoded,
    min_support=0.02,  # Increased from 0.01 to 0.02
    use_colnames=True,
    max_len=10,  # Limit maximum length of itemsets
    low_memory=True  # Use memory-efficient implementation
)

# Add length information
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))

# Display results
print("\nTop 10 frequent itemsets by support:")
print("\nShape of frequent itemsets:", frequent_itemsets.shape)
display(frequent_itemsets.nlargest(10, 'support'))

# Print memory usage
print("\nMemory usage of frequent_itemsets:", 
      frequent_itemsets.memory_usage(deep=True).sum() / 1024**2, "MB")
```

## Association Rules Generation
Generate and display association rules with minimum confidence of 50%, sorted by lift value.

```python
# Generate association rules with minimum confidence of 0.5
rules = association_rules(
    frequent_itemsets,  # Pass directly as first positional argument
    metric="confidence",
    min_threshold=0.5,
    num_itemsets=243,
)

# Sort rules by lift value
rules = rules.sort_values('lift', ascending=False)

# Display top rules and shape information
print("Number of rules generated:", len(rules))
print("\nTop 10 association rules by lift:")
display(rules.head(10))

# Show rule metrics summary
print("\nRule metrics summary:")
display(rules[['support', 'confidence', 'lift']].describe())
```

## Support-Confidence Visualization
Scatter plot showing the relationship between support and confidence metrics for generated rules.

```python
# Visualize support vs confidence
plt.figure(figsize=(10, 6))
plt.scatter(rules['support'], rules['confidence'], alpha=0.5)
plt.xlabel('Support')
plt.ylabel('Confidence')
plt.title('Support vs Confidence in Association Rules')
plt.show()
```

## Product-Specific Rules
Define and demonstrate a utility function to find association rules for specific products.

```python
def find_rules_for_product(rules_df, product_name):
    """
    Find all rules that involve a specific product either as antecedent or consequent
    """
    # Find rules where product is in antecedents
    antecedent_rules = rules_df[rules_df['antecedents'].apply(lambda x: product_name in [str(i) for i in x])]
    
    # Find rules where product is in consequents
    consequent_rules = rules_df[rules_df['consequents'].apply(lambda x: product_name in [str(i) for i in x])]
    
    # Combine both sets of rules
    combined_rules = pd.concat([antecedent_rules, consequent_rules]).drop_duplicates()
    
    return combined_rules.sort_values('lift', ascending=False)

# Find the most common product
popular_product = df['Description'].mode()[0]
print(f"Most frequent product: {popular_product}")

# Show sample of products to choose from
print("\nSample of available products:")
display(df['Description'].value_counts().head())

# Try with the most frequent product
rules_for_product = find_rules_for_product(rules, popular_product)
if len(rules_for_product) > 0:
    print(f"\nFound {len(rules_for_product)} rules involving '{popular_product}':")
    display(rules_for_product.head())
else:
    print(f"\nNo rules found for '{popular_product}'. Let's try with a sample product.")
    # Try with a sample product from the rules
    sample_product = list(next(iter(rules['antecedents'])))[0]
    print(f"\nTrying with sample product: {sample_product}")
    rules_for_product = find_rules_for_product(rules, sample_product)
    display(rules_for_product.head())
```

## Metric Distributions
Visualize the distribution of key metrics (support, confidence, and lift) across all generated rules.

```python
# Analyze rule metrics distribution
plt.figure(figsize=(12, 4))

plt.subplot(131)
plt.hist(rules['support'], bins=20)
plt.title('Support Distribution')
plt.xlabel('Support')

plt.subplot(132)
plt.hist(rules['confidence'], bins=20)
plt.title('Confidence Distribution')
plt.xlabel('Confidence')

plt.subplot(133)
plt.hist(rules['lift'], bins=20)
plt.title('Lift Distribution')
plt.xlabel('Lift')

plt.tight_layout()
plt.show()
```

## Summary and Export
Generate summary statistics of the analysis and export results to CSV files for further use.

```python
# Generate insights summary
print("\nMarket Basket Analysis Insights:")
print(f"Total number of transactions analyzed: {len(transactions)}")
print(f"Total number of unique products: {len(te.columns_)}")
print(f"Number of frequent itemsets found: {len(frequent_itemsets)}")
print(f"Number of association rules generated: {len(rules)}")
print(f"\nTop 5 products by frequency in transactions:")
display(df['Description'].value_counts().head())

# Save results to CSV
rules.to_csv('../data/association_rules.csv', index=False)
frequent_itemsets.to_csv('../data/frequent_itemsets.csv', index=False)

print("\nResults have been saved to CSV files in the data directory.")
```
